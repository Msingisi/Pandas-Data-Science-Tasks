```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('coffee.csv')
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.lower()
```

```python
columns_of_interest = ['reviews']
df.dropna(subset=columns_of_interest, inplace=True)
```

```python
df['dine_in_option'] = df['dine_in_option'].fillna('False')
```

```python
df['takeout_option'] = df['takeout_option'].fillna('False')
```

```python
df.head(50)
```

# Task1: What is the most common place type in this local market?

```python
df.columns
```

```python
# Create a bar plot
place_counts = df['place_type'].value_counts()
place_counts.plot(kind='bar')
plt.xlabel('Place Type')
plt.ylabel('Number of Places')
plt.title('Counts of Places Types in the market')
plt.xticks(rotation=0)
plt.show()
```

# Task 2: How does the range in number of reviews differ across all shops?

```python
# Create a histogram
plt.hist(df['reviews'])
plt.xlabel('Reviews')
plt.ylabel('Count')
plt.title('Range in number of reviews')
plt.show()
```

```python
#remove the outlier 
df_reviews_outlier = df[df['reviews'] < 5000]
```

```python
# Create a histogram (without the outlier)
plt.hist(df_reviews_outlier['reviews'])
plt.xlabel('Reviews')
plt.ylabel('Count')
plt.title('Range in number of reviews')
plt.show()
```

# Task 3: How does the number of reviews vary across each place type? 

```python
# Create a boxplot
plt.figure(figsize=(8, 6))
plt.boxplot([df[df['place_type'] == place]['reviews'] for place in df['place_type'].unique()],
            labels=df['place_type'].unique())
plt.xlabel('Place Type')
plt.ylabel('Reviews')
plt.title('Range in number of reviews by place type')
plt.grid(axis='y')
plt.show()
```

```python
# Create a boxplot (without the outlier)
plt.figure(figsize=(8, 6))
plt.boxplot([df_reviews_outlier[df_reviews_outlier['place_type'] == place]['reviews'] for place in df_reviews_outlier['place_type'].unique()],
            labels=df_reviews_outlier['place_type'].unique())
plt.xlabel('Place Type')
plt.ylabel('Reviews')
plt.title('Range in number of reviews by place type')
plt.grid(axis='y')
plt.show()
```
