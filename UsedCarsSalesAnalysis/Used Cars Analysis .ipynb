# Task 1: How the Sales Differ Between Hybrid Cars and Other Fuel Types?

```python
# Create a bar plot
type_counts = df['fuelType'].value_counts()
plt.bar(type_counts.index, type_counts.values, color='grey')
plt.xlabel('Fuel Type')
plt.ylabel('Used Cars Sold')
plt.title('Used Cars sold By Fuel Type')
plt.xticks(rotation=0)
plt.show()
```

```python
# Create a bar plot of Hybrid car's only 
df_hybrid = df.loc[df['fuelType'] == 'Hybrid']
model_counts = df_hybrid['model'].value_counts()
plt.bar(model_counts.index, model_counts.values, color='grey')
plt.xlabel('Car Model')
plt.ylabel('Used Cars Sold')
plt.title('Used Cars sold By Model (Hybrid Cars Only)')
plt.xticks(rotation=0)
plt.show()
```

```python
# The Distribution of price
plt.hist(df['price'], color='grey', bins=100)
plt.xlabel('Sale Price')
plt.ylabel('Number of Used Cars Sold')
plt.title('The Distribution of price')
plt.show()
```

```python
# Create a boxplot
df_outliers = df[df['price'] < 35000]
plt.figure(figsize=(8, 6))
plt.boxplot([df_outliers[df_outliers['fuelType'] ==place]['price'] for place in df_outliers['fuelType'].unique()],
            labels=df_outliers['fuelType'].unique(), patch_artist=True, boxprops={'facecolor': 'grey'})
plt.xlabel('Fuel Type')
plt.ylabel('Price')
plt.title('The Distribution of price By Fuel Type')
plt.grid(axis='y')
plt.show()
```

# Task 2: How to Make the Hybrid more Attractive?

```python
# Create a boxplot for Hybrid Cars
plt.figure(figsize=(8, 6))
sns.boxplot(x='model', y='price', data=df_hybrid, palette='gray')
plt.xlabel('Model')
plt.ylabel('Price')
plt.title('Range Of Price By Used Car Model (Hybrid Cars Only)')
plt.grid(axis='y')
plt.show()
```
